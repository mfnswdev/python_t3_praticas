<h1>Exercício 2: Pesquisa sobre persistência de dados</h1>

```python
import pickle

class ToDoList:
    def __init__(self):
        self.tarefa = []

    def lista_tarefas(self):
        print("Lista de tarefas:")
        for tarefa in self.tarefa:
            status = "[x]" if tarefa["completa"] else "[ ]"
            print(f"{tarefa['id']}. {tarefa['descrição']} {status}")

    def registrar_tarefa(self):
        descrição = input("Digite a descrição da tarefa: ").capitalize()
        tarefa = {"id": len(self.tarefa) + 1, "descrição": descrição, "completa": False}
        self.tarefa.append(tarefa)
        print("Tarefa registrada!!!")

    def marcar_tarefa_completa(self):
        tarefa_id = int(input("Digite o ID da tarefa: "))
        for tarefa in self.tarefa:
            if tarefa["id"] == tarefa_id and not tarefa["completa"]:
                tarefa["completa"] = True
                self.tarefa.remove(tarefa)
                self.tarefa.insert(0, tarefa)
                print("Tarefa marcada como concluída!")
                return
        print("Tarefa não encontrada ou já concluída.")

    def editar_tarefa(self):
        tarefa_id = int(input("Digite o ID da tarefa: "))
        for tarefa in self.tarefa:
            if tarefa["id"] == tarefa_id:
                new_description = input("Digite a nova descrição da tarefa: ").capitalize()
                tarefa["descrição"] = new_description
                print("Tarefa editada!")
                return
        print("Tarefa não encontrada.")

    def salvar_tarefa(self, filename):
        with open(filename, "wb") as file:
            pickle.dump(self.tarefa, file)
        print("Lista de tarefas salva com sucesso!")

    def carregar_tarefa(self, filename):
        try:
            with open(filename, "rb") as file:
                self.tarefa = pickle.load(file)
            print("Lista de tarefas carregada com sucesso!")
        except FileNotFoundError:
            print("Arquivo não encontrado. Nenhuma lista de tarefas encontrada.")

    def run(self):
        while True:
            print("\n===== ToDoList =====")
            print("1. Listar tarefas")
            print("2. Registrar nova tarefa")
            print("3. Marcar tarefa como concluída")
            print("4. Editar tarefa")
            print("5. Salvar tarefas em arquivo")
            print("6. Carregar tarefas de arquivo")
            print("0. Sair")

            choice = input("Escolha uma opção: ")

            if choice == "1":
                self.lista_tarefas()
            elif choice == "2":
                self.registrar_tarefa()
            elif choice == "3":
                self.marcar_tarefa_completa()
            elif choice == "4":
                self.editar_tarefa()
            elif choice == "5":
                filename = input("Digite o nome do arquivo para salvar as tarefas: ")
                self.salvar_tarefa(filename)
            elif choice == "6":
                filename = input("Digite o nome do arquivo para carregar as tarefas: ")
                self.carregar_tarefa(filename)
            elif choice == "0":
                break
            else:
                print("Opção inválida. Tente novamente.")


# Instanciar e executar o aplicativo
app = ToDoList()
app.run()
```
