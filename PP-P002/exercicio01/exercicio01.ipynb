```python
class ToDoList:
    def __init__(self):
        self.tarefas = []

    def lista_tarefas(self):
        print("Lista de tarefas:")
        for tarefa in self.tarefas:
            status = "[x]" if tarefa["completa"] else "[ ]"
            print(f"{tarefa['id']}. {tarefa['descrição']} {status}")

    def registrar_tarefa(self):
        descrição = input("Digite a descrição da tarefa: ").capitalize()
        tarefa = {"id": len(self.tarefas) + 1, "descrição": descrição, "completa": False}
        self.tarefas.append(tarefa)
        print("Tarefa registrada!!!")

    def marcar_tarefa_completa(self):
        task_id = int(input("Digite o ID da tarefa: "))
        for tarefa in self.tarefas:
            if tarefa["id"] == task_id and not tarefa["completa"]:
                tarefa["completa"] = True
                self.tarefas.remove(tarefa)
                self.tarefas.insert(0, tarefa)
                print("Tarefa marcada como concluída!")
                return
        print("Tarefa não encontrada ou já concluída.")

    def editar_tarefa(self):
        task_id = int(input("Digite o ID da tarefa: "))
        for tarefa in self.tarefas:
            if tarefa["id"] == task_id:
                new_description = input("Digite a nova descrição da tarefa: ").capitalize()
                tarefa["descrição"] = new_description
                print("Tarefa editada!")
                return
        print("Tarefa não encontrada.")

    def run(self):
        while True:
            print("\n===== ToDoList =====")
            print("1. Listar tarefas")
            print("2. Registrar nova tarefa")
            print("3. Marcar tarefa como concluída")
            print("4. Editar tarefa")
            print("0. Sair")

            choice = input("Escolha uma opção: ")

            if choice == "1":
                self.lista_tarefas()
            elif choice == "2":
                self.registrar_tarefa()
            elif choice == "3":
                self.marcar_tarefa_completa()
            elif choice == "4":
                self.editar_tarefa()
            elif choice == "0":
                break
            else:
                print("Opção inválida. Tente novamente.")


# Instanciar e executar o aplicativo
app = ToDoList()
app.run()
```
