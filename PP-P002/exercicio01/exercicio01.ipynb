<h1>Exercício 1: Utilizando estruturas de controle de fluxo e listas.</h1>
<p>Para controlar os compromissos do dia a dia podemos utilizar um uma lista de tarefas. Com alguns dos recursos básicos da linguagem que já conhecemos podemos implementar o aplicativo ToDoList que permita:</p>

```python
lista_tarefas = []

def listar_tarefas():
    if not lista_tarefas:
        print("Não tem Tarefas registradas!!")
        return
    print("Lista de tarefas: ")
    for id, tarefa in enumerate(lista_tarefas, start=1):
        print(f"{id}. {tarefa}")
        
def registrar_nova_tarefa():
    descricao_tarefa = input("Informe a descrição da tarefa: ").capitalize() ## capitalizer para deixar a string no padrao
    tarefa = f"{len(lista_tarefas)+1}. {descricao_tarefa} [ ]"
    lista_tarefas.append(tarefa)
    print("Tarefa registrada com sucesso!")
    
def check_tarefa(): 
    listar_tarefas()
    identificador = int(input("Informe o id da tarefa que voce deseja marcar como concluída? "))
    isIdentificarInList = 1 <= identificador <= len(lista_tarefas)
    if isIdentificarInList:
        tarefa = lista_tarefas.pop(identificador - 1)
        tarefa = tarefa.replace("[ ]", "[X]")
        lista_tarefas.insert(0, tarefa)
        print("A Tarefa foi marcada como concluida !!")
    else:
        print("ERRO: O Id informado não existe na lista !!!")

def editar_tarefa():
    listar_tarefas()
    identificador = int(input("Informe o id da tarefa que voce deseja editar? "))
    isIdentificarInList = 1 <= identificador <= len(lista_tarefas)
    if isIdentificarInList:
        tarefa = lista_tarefas.pop(identificador - 1)
        nova_tarefa = input("Qual é a nova descrição da tarefa? ")
        tarefa = f"{identificador}. {nova_tarefa} [ ]"
        lista_tarefas.insert(0, tarefa)
        print("Tarefa editada com sucesso!")
    else:
        print("ERRO: O Id informado não existe na lista !!!")
    
def main():
    while True:
        print("\nLista de Tarefas:")
        print("1. Listar tarefas")
        print("2. Registrar nova tarefa")
        print("3. Marcar tarefa como concluída")
        print("4. Editar tarefa")
        print("0. Sair")
        escolha = input("Escolha uma opção: ")
        
        if escolha == "1":
            listar_tarefas()
        elif escolha == "2":
            registrar_nova_tarefa()
        elif escolha == "3":
            check_tarefa()
        elif escolha == "4":
            editar_tarefa()
        elif escolha == "0":
            print("Saindo da Lista de Tarefas!!")
            break
        else:
            print("Opção inválida, tente novamente!!")
        
        input("Pressione Enter para continuar...")
        
if __name__ == "__main__":
    main()
```
